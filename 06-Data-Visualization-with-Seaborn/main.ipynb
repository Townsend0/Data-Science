```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.heatmap(flights.pivot_table(index = 'year', columns = 'month', values = 'passengers'),
linecolor = 'white', lw = 1, cmap = 'magma')
```

```python
sns.clustermap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'), cmap = 'magma')
```

```python
grid = sns.PairGrid(iris)
grid.map_upper(sns.scatterplot)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot)
```

```python
grid = sns.FacetGrid(tips, row = 'smoker', col = 'time')
grid.map(sns.histplot, 'total_bill', 'tip')
```

```python
sns.lmplot(tips, x = 'total_bill', y = 'tip', hue = 'sex', aspect = 2, markers = ['s', 'D'])
```
