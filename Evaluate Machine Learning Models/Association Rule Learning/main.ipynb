```python
from apyori import apriori
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.set_style('darkgrid')
df = pd.read_csv('./Market_Basket_Optimisation.csv', header = None)
```

```python
results = apriori(list(list(b for b in a if not pd.isna(b)) for a in df.values), min_lift = 3,
min_support =  3 * 7 / df.index.size, min_confidence = 0.2, min_length = 2, max_length = 2)
```

```python
data = pd.DataFrame(columns = ['First Item', 'Second Item', 'Support', 'Confidence', 'Lift'])
```

```python
for rule in enumerate(results):
    data.loc[rule[0]] = {'First Item': tuple(rule[1][2][0][0])[0], 'Second Item': tuple(rule[1][2][0][1])[0], 
    'Support': f'{round(rule[1][1] * 100, 2)}%', 'Confidence': f'{round(rule[1][2][0][2] * 100, 2)}%',
    'Lift': f'{round(rule[1][2][0][3] * 100, 2)}%'}

data.sort_values('Lift', ascending = False)
```
